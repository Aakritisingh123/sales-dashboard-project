FEATURE ENGINEERING

```python
df['Sales Category'] = pd.cut(df['Sales'], bins = [0,100,500,1000,5000,25000], labels= ['Low','Medium','High','Very High','Premium'])
```

OUTLIER DETECTION

```python
def check_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_limit) | (df[column] > upper_limit)]

    return outliers, lower_limit, upper_limit
```

1. Sales Outliers

```python
sales_outliers, sales_low, sales_high = check_outliers('Sales')
print("Sales outliers:", len(sales_outliers))
print("Lower limit:", sales_low)
print("Upper limit:", sales_high)
```

```python
sales_outliers[['Category','Sub-Category','Sales','Profit']].head(20)
```

```python
df['Sales Outlier'] = np.where((df['Sales'] < sales_low) | (df['Sales'] > sales_high),'Yes','No')
```

```python
df.head()
```

2. Profit Outliers

```python
profit_outliers, profit_low, profit_high = check_outliers('Profit')
print("Profit outliers:", len(profit_outliers))
print("Lower limit:", profit_low)
print('Upper limit:', profit_high)
```

```python
profit_outliers, profit_low, profit_high = check_outliers("Profit")

df["Profit Outlier"] = np.where(
    (df["Profit"] < profit_low) |
    (df["Profit"] > profit_high),
    "Yes",
    "No"
)
```

```python
df.head()
```

3. Qunatity Outliers

```python
quantity_outliers, q_low, q_high = check_outliers("Quantity")

print("Quantity outliers:", len(quantity_outliers))
```

```python
quantity_outliers, q_low, q_high = check_outliers("Quantity")

df["Quantity Outlier"] = np.where(
    (df["Quantity"] < q_low) |
    (df["Quantity"] > q_high),
    "Yes",
    "No"
)
```
